DATA VISUALIZATION

```python
cols = ['1990 [YR1990]', '2000 [YR2000]', '2013 [YR2013]',
       '2014 [YR2014]', '2015 [YR2015]', '2016 [YR2016]', '2017 [YR2017]',
       '2018 [YR2018]', '2019 [YR2019]', '2020 [YR2020]', '2021 [YR2021]',
       '2022 [YR2022]']
```

```python
import matplotlib.pyplot as plt

for i in cols:
    fig = plt.figure(figsize=(10,6))
    plt.hist(df[i], color='#1f77b4', bins=10)
    plt.xlabel(i)
    plt.show()
    
```

Plotting barplot for each year

```python
#extract the years (cols) from the dataframe
years = df.columns[1:]
#sum the values for each year across all countries
total_values = df[years].sum()
#plotting the bar chart
plt.figure(figsize=(30,30))
plt.barh(years, total_values, color = '#191970')
plt.xlabel('Total Values')
plt.ylabel('Year', size=20)
plt.title('Total values per year', size=20)
plt.show()
```

```python
country_by_2000=df.sort_values(by='2000 [YR2000]').head(10)
country_by_2000
```

```python
country_by_2000_t = country_by_2000.set_index('Country Name').T
for country_name, data_values in country_by_2000_t.iterrows():
    fig = plt.figure(figsize=(3,3))
    sns.barplot(x=data_values.index, y=data_values.values)
    plt.xlabel('Countries')
    plt.ylabel('Data values')
    plt.title(f"(country_name).Data values from 2021 to 2022")
    plt.xticks(rotation = 90)
    plt.show()
    
```
